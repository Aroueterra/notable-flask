```python
def unique(sequence):
    seen = set()
    return [x for x in sequence if not (x in seen or seen.add(x))]

import json
import re
import copy
from midi.vocabulary import *
voc_file = "vocabulary_semantic2.txt"
pred_file = "aside/predictions.txt"
edge = "vocab3.txt"
dict_file = open(voc_file,'r')
pred_file = open(pred_file,'r')
edge = open(edge,'r')
semantic = dict_file.read().splitlines()
pred_file = pred_file.read().splitlines()
edge = edge.read().splitlines()
fnd_str = re.compile(r'((note|gracenote|rest|multirest)(\-)(\S)*)').findall(str(semantic))
music_str = [i[0] for i in fnd_str]
fnd_notes = [re.compile(r'(([A-G](b|#)?[1-6])|rest)').findall(note) for note in music_str]
notes = [m[0][0] for m in fnd_notes]
fnd_durs = [re.compile(r'((\_|\-)([a-z]|[0-9])+(\S)*)+').findall(note) for note in music_str]
durs = [i[0][0][1:].split('_') for i in fnd_durs]

jsondata = {}
#Setup top-level
note={}
gracenote={}
rest={}
multirest={}
barline={}
clef={}
tie={}
keySignature={}
timeSignature={}
barline={}
tie={}
toplevel=[]
toplevel.extend((note,gracenote,rest,multirest,barline,clef,tie,keySignature,timeSignature))

#Setup nested
note_type={}
note_body={}
notation={}
term = []

for val in music_str:
    term.append(val.split('-')[0])
idx = 0
prev_term = None
terms = unique(term)
for value in terms:
    newType = dict()
    newNotation = dict()
    notation = dict()
    notation['name'] = value + ': ' 
    notation['definition'] = 'placeholder'
    note_type = dict()
    check = term[idx]
    while value == check:
        note_body = dict()
        note_body['name'] = notes[idx]
        if value == "gracenote" or value == "note":
#             print(type(value))
#             print((value))
#             print(type(notes[idx]))
#             print((notes[idx]))
            pitch_freq = FREQ[str(notes[idx])]
            note_body['definition'] = str(notes[idx]) + " has a frequency of " + str(pitch_freq)
        else:
            note_body['definition'] = "definition of " + notes[idx]    
        note_type[notes[idx]] = copy.deepcopy(note_body)
        idx+=1
        if idx < len(term):
            check = term[idx]
        else:
            break
    newType = copy.deepcopy(note_type)
    notation['type'] = newType
    newNotation = copy.deepcopy(notation)
    jsondata[value] = newNotation   
    
idx = 0
shortterm = []
shortvalue = []
for val in edge:
    shortterm.append(val.split('-')[0])
    shortvalue.append(val.split('-')[1])
shortterms=unique(shortterm)

for value in shortterms:
    newType = dict()
    newNotation = dict()
    notation = dict()
    notation['name'] = value + ': ' 
    notation['definition'] = 'placeholder'
    note_type = dict()
    check = shortterm[idx]
    #print(value + " " + str(idx) + " " + shortterm[idx])
    while value == check:
        note_body = dict()
        note_body['name'] = shortvalue[idx] 
        note_body['definition'] = "definition of " + shortvalue[idx] 
        note_type[shortvalue[idx]] = copy.deepcopy(note_body)
        #print(value + " " + str(idx) + " V: " + shortvalue[idx] + " T: " + shortterm[idx] )
        idx+=1
        if idx < len(shortterm):
            check = shortterm[idx]
        else:
            break
    newType = copy.deepcopy(note_type)
    notation['type'] = newType
    newNotation = copy.deepcopy(notation)
    jsondata[value] = newNotation
    
    barline['name'] = 'Barline: ' 
    barline['definition'] = 'placeholder'
    jsondata['barline'] = barline
    tie['name'] = 'Tie: ' 
    tie['definition'] = 'placeholder'
    jsondata['tie'] = tie

print("Process complete.")

with open('data.json', 'w', encoding='utf-8') as f:
    json.dump(jsondata, f, ensure_ascii=False, indent=4)
```
